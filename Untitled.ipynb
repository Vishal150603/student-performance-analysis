```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv(
    r"data.csv\Expanded_data_with_more_features.csv"
)

print(df.head())
```

```python
df.info()
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("gender distribution")
plt.show()
```

# from the above data we have analysed that:
# the number of females in te data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot = True)
plt.title("Relationship b/w parents education and student's score")
plt.show()
```

# from the abovw chart we have conclude that the education of the parents have a good impact

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb1,annot = True)
plt.title("Relationship b/w parents marital status and student's score")
plt.show()
```

 # from the above chart we have conclude that there is neglegible impact on the student's score due to their parent's marital status
